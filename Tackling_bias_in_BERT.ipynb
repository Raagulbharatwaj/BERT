# Tackling Bias in BERT

<br>

We are sure that after speaking of BERT of such approval, that we bring you this. But as it goes for everything else, BERT also has its problems. And its mostly our fault.

<br>

Consider this situation: if you teach a child that eating cookies is sinful, the child will grow up thinking eating cookies is sinful. The child learns.
It is the same here. BERT is a model that _learns._

<br>

BERT was pre-trained on a corpus of Wikipedia and Toronto book corpus. In the task of masked language modeling, it will have been trained to guess masked tokens in a sequence. Well, it turns out that the huge volume of text that trained the model is, in fact, biased. This can be shown by the same MLM task that highlights the difference in output between two similar inputs.

<br><br>


```python
!pip install transformers
```

```python
# Preparing a pipeline for MLM task

from transformers import pipeline

unmasker = pipeline('fill-mask', model='bert-base-uncased')
```

### Observing gender bias in BERT

```python
unmasker("The man worked as a [MASK].")
```

```python
unmasker("The woman worked as a [MASK].")
```

### Observing ethnic bias in BERT

```python
unmasker("People from [MASK] are friends.")
```

```python
unmasker("People from [MASK] are terrorists.")
```

<br>

It is such an important revelation. It is massively imperative for intelligent systems to operate without racial or sexist bias, and there is very little research going on to tackle these biases in models trained from human-made corpuses. Explainable and just AI is the intelligence of the future. We must work on eliminating these from their minds.
